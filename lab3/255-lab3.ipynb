```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn import svm
```

```python
df = pd.read_csv('../input/fetal-health-classification/fetal_health.csv')
df.head()
```

```python
X = df.drop(['fetal_health'], axis=1)
Y = df["fetal_health"]
print(X)
print(Y)
```

```python
from sklearn import svm
# Metrics
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

degreeArr = [3,4,5,6,8]
CArr = [0.7,0.8,0.9,1, 1.5]
Accuracy = []
Precision = []
Recall = []
F1Score =[]

for x in range(len(degreeArr)): #This assumes degreeArr and CArr are the same length
  model = svm.SVC(kernel='poly', degree=degreeArr[x], C=CArr[x]).fit(X, Y)
  poly_svc_test_result = model.predict(X);
  print(model)
    
  accuracy = accuracy_score(poly_svc_test_result, Y)
  print("The SVC with polynomial (degree 3) kernel Accuracy is = ", accuracy*100, " %")
  matrix = confusion_matrix(Y, poly_svc_test_result)
  precision = matrix[0][0]/(matrix[0][0]+matrix[1][0])
  recall = matrix[0][0]/(matrix[0][0]+matrix[0][1])
  f1score = (matrix[0][0]*2)/(matrix[0][0]*2+matrix[1][0]+matrix[0][1])
  print("The SVC with polynomial (degree 3) kernel Precision is = ", precision)
  print("The SVC with polynomial (degree 3) kernel Recall is = ", recall)
  print("The SVC with polynomial (degree 3) kernel F1-score is = ", f1score)
  report = classification_report(Y, poly_svc_test_result)
  print(report)  
  Accuracy.append(accuracy)
  Precision.append(precision)
  Recall.append(recall)
  F1Score.append(f1score)

#plotting
plt.plot(CArr, Accuracy)
plt.xlabel('C Value')
plt.ylabel('Accuracy')
plt.title('Accuracy score as C increases')
plt.grid(linestyle='dotted')
plt.savefig('accuracy.png')
plt.show()

plt.plot(CArr, Precision)
plt.xlabel('C Value')
plt.ylabel('Precision')
plt.title('Precision score as C increases')
plt.grid(linestyle='dotted')
plt.savefig('precision.png')
plt.show()


plt.plot(CArr, Recall)
plt.xlabel('C Value')
plt.ylabel('Recall')
plt.title('Recall score as C increases')
plt.grid(linestyle='dotted')
plt.savefig('recall.png')
plt.show()

plt.plot(CArr, F1Score)
plt.xlabel('C Value')
plt.ylabel('F1Score')
plt.title('F1Score score as C increases')
plt.grid(linestyle='dotted')
plt.savefig('F1Score.png')
plt.show()




# apply different kernel functions
#svc = svm.SVC(kernel='linear', C=CArr).fit(X, Y)
#rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=CArr).fit(X, Y)
#poly_svc = svm.SVC(kernel='poly', degree=3, C=CArr).fit(X, Y)
#lin_svc = svm.LinearSVC(C=C).fit(X, Y)
```
